```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns',None)

df  = pd.read_excel("../data/Pakistani_Foods_Nutrition_FYP.xlsx")
df1 = pd.read_csv("../data/Foods_data.csv")
df2 = pd.read_csv("../data/Indian_Food_Nutrition_Processed.csv")
```

```python
df.rename(columns={"Dish Name (Urdu/Common Name)": "Dish Name"}, inplace=True)

```

```python
df1["name"] = df1["name"].str.split("Recipe").str.get(0).str.strip().str.split("(").str.get(0).str.strip()   # Dulce De Leche (Boiled Condensed Milk)   # Pakistani Aloo Palak

name_issue = {"Dulce De Leche (Boiled Condensed Milk)":"Boiled Condensed Milk","Pakistani Aloo Palak":"Aloo Palak"}
df1["Dish Name"] = df1["name"].replace(name_issue,inplace=True)
```

```python
import ast
df1["Calories (kcal)"]  = df1["nutritions"].apply(lambda x: ast.literal_eval(x)["Calories"].replace(" kcal", "").strip())
```

```python
df1["nutritions"][0]
```

```python
df1["nutritions"] = df1["nutritions"].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

df1["Carbohydrates (g)"] = df1["nutritions"].apply(lambda x: x.get("Carbohydrates", "0").split()[0])
df1["Protein (g)"]       = df1["nutritions"].apply(lambda x: x.get("Protein", "0").split()[0])
df1["Fats (g)"]          = df1["nutritions"].apply(lambda x: x.get("Fat", "0").split()[0])
df1["Free Sugar (g)"]    = df1["nutritions"].apply(lambda x: x.get("Sugar", "0").split()[0])
df1["Fibre (g)"]         = df1["nutritions"].apply(lambda x: x.get("Fiber", "0").split()[0])
df1["Sodium (mg)"]       = df1["nutritions"].apply(lambda x: x.get("Sodium", "0").split()[0])
df1["Calcium (mg)"]      = df1["nutritions"].apply(lambda x: x.get("Calcium", "0").split()[0])
df1["Iron (mg)"]         = df1["nutritions"].apply(lambda x: x.get("Iron", "0").split()[0])
df1["Vitamin C (mg)"]    = df1["nutritions"].apply(lambda x: x.get("Vitamin C", "0").split()[0])
```

```python
df1  = df1.iloc[:,9:]
```
